```python
import re

raceCategories = ["Native American","Asian","Black","White","Latino", "Mixed","Other"]
refdf["Race_Ethnicity"] = refdf["Race_Ethnicity"].str.lower()


for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[\s/-]",race)
    #if len(x) == 1:
    x = re.sub("multi|mixed", "Mixed", x[0])        
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
        

refdf["Race_Ethnicity"]
```

```python
for index,race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[/-]",race)
   # if len(x) == 1:     
    x = re.sub("philippine|hispanic|mexican|mexicano|puerrto|puerta rican|spanish|latin.", "Latino", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
    
refdf["Race_Ethnicity"]
```
