```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import the dataset

```python
refdf = pd.read_csv("ReFresh.csv")
refdf.head()
```

```python
columns= list(refdf.columns)
columns
```

# Sort out the columns to identify the useful ones

```python
postsuf = "Post"
presuf = "Pre"
pre = []
post = []
wout = []
for variable in columns:
    if postsuf in variable:
        post.append(variable)
    elif presuf in variable:
    	pre.append(variable)
    else:
        wout.append(variable)

print("Pre: {}".format(len(pre)),
"Post: {}".format(len(post)),
"Without: {}".format(len(wout)))

print(pre)
print(post)
print(wout)    
```

# Match Before and After ReFresh Center Scores

```python
prepostcolumns = []

for col in pre:
    for col2 in post:
        x = col.split('Pre')
        y = col2.split('Post')
       # print(x," ",y)
        if x == y:
            prepostcolumns.append(col)
            prepostcolumns.append(col2)
            
prepostcolumns
```

# Add important Columns (like Gender,Ethnicity) to the Before and After Scores

```python
for col in wout:
    index = wout.index(col)
    prepostcolumns.insert(index,col)
    
print(prepostcolumns)
```

# Clean up the ethnicity column using multiple rounds of Regular Expressions

```python
import re

raceCategories = ["Native American","Asian","Black","White","Latino", "Mixed","Other"]
refdf["Race_Ethnicity"] = refdf["Race_Ethnicity"].str.lower()


for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[\s/-]",race)
    #if len(x) == 1:
    x = re.sub("multi|mixed", "Mixed", x[0])        
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
        

refdf["Race_Ethnicity"]
```

```python
for index,race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[/-]",race)
   # if len(x) == 1:     
    x = re.sub("philippine|hispanic|mexican|mexicano|puerrto|puerta rican|spanish|latin.", "Latino", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
    
#refdf["Race_Ethnicity"]
```

```python
for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[\s/-]",race)
   # if len(x) == 1:     
    x = re.sub("jamican|africanamerican|blk|blhuman|black|african", "Black", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
```

```python
for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[/-]",race)
    #if len(x) == 1:     
    x = re.sub("american|native|islander", "Native American", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
```

```python
for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[/-]",race)
    #if len(x) == 1:     
    x = re.sub("white|irish|italian|portugese|usa", "White", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
```

```python
for index, race in enumerate(refdf["Race_Ethnicity"]):
    x = re.split("[/-]",race)
    if len(x) == 1:     
    	x = re.sub("japanese", "Asian", x[0]);
			
    print("Before: {}".format(race),"| After: {}".format(x))
    refdf["Race_Ethnicity"][index] = x
```

```python

#raceClean= set(refdf["Race_Ethnicity"]).difference(raceCategories)

#raceCleanRows = refdf["Race_Ethnicity"].isin(raceClean)

for ind, val in enumerate(refdf["Race_Ethnicity"]):
    if val not in raceCategories:
        refdf["Race_Ethnicity"][ind] = "Other"
     
        
        
        
refdf["Race_Ethnicity"].value_counts()
```
